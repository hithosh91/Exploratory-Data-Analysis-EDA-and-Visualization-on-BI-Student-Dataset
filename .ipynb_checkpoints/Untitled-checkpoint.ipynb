```python
df['prevEducation'].unique()
```

```python
def updateeducation(x):
    if x =='Masters':
      return 'Masters'
    elif  x=="Diploma" or x=='diploma' or x== 'Diplomaaa' or x== 'DIPLOMA':
      return "Diploma"
    elif x=="HighSchool" or x=='High School':
      return "HighSchool"
    elif x== 'Bachelors' or x=='Barrrchelors' :
      return 'Bachelors'
    elif x== 'Doctorate':
        return 'Doctorate'

df['prevEducation']=df['prevEducation'].apply(updateeducation)
        
```

```python
df['prevEducation'].unique()
```

```python
#checking any missinng values
df.isnull().sum()
```

```python
# update missing values
df['Python'] = df['Python'].fillna(df['Python'].mean())
```

```python
numerical_data=['Age','entryEXAM','studyHOURS','Python','DB']
```

```python
df[numerical_data].hist(bins=10, figsize=(12,8), color='skyblue', edgecolor='black')
plt.suptitle("Distribution of Numeric Features")
plt.show()
```

```python
category_analsis=['gender','country','residence','prevEducation']
for col in category_analsis:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette="Set2")
    plt.title(f"Count of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
corr = df[numerical_data].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Python scores by gender
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="gender", y="Python", palette="Pastel1")
plt.title("Python Scores by Gender")
plt.show()

# DB scores by previous education
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="prevEducation", y="DB", palette="Pastel2")
plt.title("DB Scores by Previous Education")
plt.xticks(rotation=45)
plt.show()
```

```python
# Python scores vs study hours by gender
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="studyHOURS", y="Python", hue="gender", palette="Set1")
plt.title("Python Scores vs Study Hours by Gender")
plt.show()

# DB scores vs entry exam by previous education
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="entryEXAM", y="DB", hue="prevEducation", palette="Set2")
plt.title("DB Scores vs Entry Exam by Previous Education")
plt.show()
```
